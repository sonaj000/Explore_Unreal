```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Sample data
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Generate colors based on the frequency of points
data = np.column_stack((x, y, z))
unique, counts = np.unique(data, axis=0, return_counts=True)

# Normalize the counts to get colors between 0 and 1
min_count = np.min(counts)
max_count = np.max(counts)

if min_count != max_count:
    counts_norm = (counts - min_count) / (max_count - min_count)
else:
    # If all counts are the same, set all colors to 0
    counts_norm = np.zeros_like(counts, dtype=float)

# Scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with custom colors
ax.scatter(x, y, z, c=counts_norm, cmap='viridis')

# Set labe


```

```python
def count_cumulative_unique(series):
    seen = set()
    count = []
    for item in series:
        seen.add(item)
        count.append(len(seen))
    return count
```

```python
unique_counts = count_cumulative_unique(df['CellString'])
plt.plot(unique_counts)
plt.xlabel("number of fastticks")
plt.ylabel("number of unique cells")
#plt.yscale("log")

#maybe plot this 3d? and as we run through it, we can show the graph changing? maybe wondering about making an interactive visual where the dev marks certain points? 
```
